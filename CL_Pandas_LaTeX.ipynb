# Exporting a dataframe into a LaTeX table

## Importing the required libraries

```python
import pandas as pd
```

## Loading the data into a dataframe

```python
df = pd.read_excel('Table.xlsx')
```

```python
df
```

### Converting the column `%` into percentages

```python
df['%'] = df['%']*100
```

```python
df
```

## Exporting the dataframe into a LaTeX table

Please refer to [pandas.DataFrame.to_latex](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_latex.html#pandas.DataFrame.to_latex).

```python
df_latex = df.to_latex(index=False, float_format='%.2f', longtable=True, decimal=',', caption='Social presence assessment', label='tab:social_presence_assessment')
```

## Saving the LaTeX table into a file

```python
with open('Table.tex', 'w', encoding='utf8', newline='\n') as file:
    file.write(df_latex)
```
