# Corpus filtering

## Requirements

### Requirements of 11th September, 2023
Talvez se certificar de que todos os tweets tenham pelo menos uma ocorrência desses termos. O twarc output inclui respostas aos tweets que não têm necessariamente os termos de busca, diferentemente do snscrape

Comando para fazer isso:
grep -E 'vacin|pfizer|astra|coronavac|janssen|jansen|covaxin|antivax|sputnik|dose|vachina|covax' arquivo_com_os_tweets > output.txt

A busca é por strings, então não é preciso usar variações que contêm a mesma string

### Requirements of 17th October, 2023
Do corpus 'text_uniq_first_view.xlsx', remover as palavras
- gripe
- hpv
- thru
- música
- prorrogada(o)
- fofura
- dupla

## Importing the required libraries

```python
import pandas as pd
```

## Loading the corpus into a dataframe

```python
df_corpus = pd.read_excel('text_uniq_first_view.xlsx')
```

```python
df_corpus.head()
```

## Filtering the corpus according to the requirements

```python
df_corpus_filtered = pd.DataFrame(columns = df_corpus.columns)
df_corpus_removed = pd.DataFrame(columns = df_corpus.columns)
for index, row in df_corpus.iterrows():
    if ('gripe' or 'hpv' or 'thru' or 'música' or 'prorrogada(o)' or 'fofura' or 'dupla') in df_corpus.loc[index, 'text_emojified']:
        df_corpus_filtered = pd.concat([df_corpus_filtered, df_corpus.loc[index].to_frame().T], ignore_index = True)
    else:
        df_corpus_removed = pd.concat([df_corpus_removed, df_corpus.loc[index].to_frame().T], ignore_index = True)
```

### Corpus - filtered

```python
df_corpus_filtered
```

### Corpus - removed

```python
df_corpus_removed
```

## Writing the results into Excel

```python
writer_filtered = pd.ExcelWriter('text_uniq_first_view_filtered.xlsx')
df_corpus_filtered.to_excel(writer_filtered, sheet_name = 'Sheet1')
writer_filtered.close()
writer_removed = pd.ExcelWriter('text_uniq_first_view_removed.xlsx')
df_corpus_removed.to_excel(writer_removed, sheet_name = 'Sheet1')
writer_removed.close()
```
